```python
# Import necessary libraries
import numpy as np #For Linear Algebra
import matplotlib.pyplot as plt #For Data Visualization
from math import sin, pi, sqrt, cos #For Mathematical Operations
```

### Linearized Solution

```python
g, l = 9.81, 4*9.81
K = sqrt(g/l) # Angular Frequency
h = 0.01 #Step Size
it = 5000 #Iterations
n = np.array(range(it+1)) #Iteration Array
t = n*h #Time-step Array
theta_0 = pi/2 #Initial postion
theta_dot_0 = 0.0 #Initial velocity
theta = np.array([theta_0]) #Array of positions
theta_dot = np.array([theta_dot_0]) #Array of velocities
# Algorithm
for i in range(1, it+1):
    theta_1 = theta_0 + h*theta_dot_0
    theta_dot_1 = theta_dot_0 - h*(K**2)*theta_1
    theta = np.append(theta, theta_1)
    theta_dot = np.append(theta_dot, theta_dot_1)
    theta_0 = theta_1
    theta_dot_0 = theta_dot_1
# Plot the position-time graph
plt.plot(t, theta)
plt.xlabel('t')
plt.ylabel('theta(t)')
plt.show()
# Plot the state-space graph
plt.plot(theta, theta_dot)
plt.xlabel('theta(t)')
plt.ylabel('omega(t)')
plt.show()
theta_linear = theta
theta_dot_linear = theta_dot
```

### Complete Non-Linear Solution

```python
g, l = 9.81, 4*9.81
K = sqrt(g/l) # Angular Frequency
h = 0.01 #Step Size
it = 5000 #Iterations
n = np.array(range(it+1)) #Iteration Array
t = n*h #Time-step Array
theta_0 = pi - 0.1 #Initial postion
theta_dot_0 = 0.0 #Initial velocity
theta = np.array([theta_0]) #Array of positions
theta_dot = np.array([theta_dot_0]) #Array of velocities
# Algorithm
for i in range(1, it+1):
    theta_1 = theta_0 + h*theta_dot_0
    theta_dot_1 = theta_dot_0 - h*(K**2)*sin(theta_1)
    theta = np.append(theta, theta_1)
    theta_dot = np.append(theta_dot, theta_dot_1)
    theta_0 = theta_1
    theta_dot_0 = theta_dot_1
# Plot the position-time graph
plt.plot(t, theta)
plt.plot(t, theta_linear)
plt.legend(["Non-Linear", "Linear"], loc ="lower right")
plt.xlabel('t')
plt.ylabel('theta(t)')
plt.show()
# Plot the state-space graph
plt.plot(theta, theta_dot)
plt.plot(theta_linear, theta_dot_linear)
plt.legend(["Non-Linear", "Linear"], loc ="lower right")
plt.xlabel('theta(t)')
plt.ylabel('omega(t)')
plt.show()
```
