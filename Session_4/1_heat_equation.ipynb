```python
# Import necessary libraries
import numpy as np #For Linear Algebra
import matplotlib.pyplot as plt #For Data Visualization
import math #For Mathematical Operations
```

```python
# Define a function
def f(x):
  return x**0.5
```

### Second-Order Partial Differential Equation

```python
alpha = 0.1 #Equation Constant
L = 10 #Length of Rod
N = 100 #Number of Divisions
dx = L/N #Length of Divisions
dt = 0.1 #Time-Step
it = 100 #Iterations
n = np.array(range(0, it+1)) #Iterations array
t = n*dt #Time-Step Vector
k = (alpha*dt)/(dx**2) #Constant
x = np.array(range(0, N+1))*dx #Position Step Vector
u = np.array([[0]*(N+1)]*(it+1), dtype=float) #Array for storing Temperature values
# Initial condition
for i in range(1, N+1):
    u[0][i] = f(i*dx)
# Algorithm
for j in range(0, it-1):
    for i in range(1, N-1):
        u[j+1][i] = u[j][i] + k*(u[j][i+1] - 2*u[j][i] + u[j][i-1])
    u[j][0] = 0 #Boundary condition - 1
    u[j][N] = 0 #Boundary condition - 2
```

```python
# Plot 1
plt.plot(x, u[0])
plt.xlabel('Position')
plt.ylabel('Temperature')
plt.show()
# Plot 2
plt.plot(x, u[3])
plt.xlabel('Position')
plt.ylabel('Temperature')
plt.show()
# Plot 3
plt.plot(x, u[5])
plt.xlabel('Position')
plt.ylabel('Temperature')
plt.show()
# Plot 4
plt.plot(x, u[7])
plt.xlabel('Position')
plt.ylabel('Temperature')
plt.show()
# Plot 5
plt.plot(x, u[9])
plt.xlabel('Position')
plt.ylabel('Temperature')
plt.show()
# Plot 6
plt.plot(x, u[10])
plt.xlabel('Position')
plt.ylabel('Temperature')
plt.show()
# Plot 7
plt.plot(x, u[25])
plt.xlabel('Position')
plt.ylabel('Temperature')
plt.show()
# Plot 8
plt.plot(x, u[50])
plt.xlabel('Position')
plt.ylabel('Temperature')
plt.show()
# Plot 9
plt.plot(x, u[75])
plt.xlabel('Position')
plt.ylabel('Temperature')
plt.show()
# Plot 10
plt.plot(x, u[100])
plt.xlabel('Position')
plt.ylabel('Temperature')
plt.show()
```
